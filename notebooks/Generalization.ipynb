```python
schema = "disease_outbreak"

# Load key verbs
with open(f"{os.environ['EVENT_HOME']}/results/schema_verbs.json") as f:
    key_verbs = set(json.load(f)[schema])

common_args = Counter()
generalized_terms = {}
related_docs = f"{os.environ['EVENT_HOME']}/results/{schema}/nyt_en_{schema}.jsonl.predargs"
with jsonlines.open(related_docs) as reader:
    for obj in reader.iter():
        for pred_args in obj['pred_args']:
            pred_idx = pred_args[2]
            arg_idx = 1 - pred_idx
            pred, arg = pred_args[pred_idx], pred_args[arg_idx].lower()
            arg = lemmatizer.lemmatize(arg, pos=wn.NOUN)
            if pred not in key_verbs or arg.lower() in stopwords:
                continue
            
            # Check if word is in the custom synset list
            # Use WordNet lookup if not
            if arg in synsets[schema]:
                syn_root = synsets[schema][arg]
            else:
                if not wn.synsets(arg, pos=wn.NOUN):
                    continue
                syn = wn.synsets(arg, pos=wn.NOUN)[0]
                if syn.hypernyms():
                    syn_root = syn.hypernyms()[0].name()
                else:
                    syn_root = syn.name()
            common_args.update([(pred, syn_root)])
            generalized_terms[syn_root] = generalized_terms.get(syn_root, set())
            generalized_terms[syn_root].add(arg)
            
            # print(f"Pred: {pred}\tArg: {arg}\t{syn_root}")
```

```python
for arg, count in common_args.most_common():
    if arg[1] != "illness.n.01":
        continue
    print(f"{arg}\t{count:,}")
```

```python
for root, args in generalized_terms.items():
    print(f"{root}\t{' '.join(args)}")
```
