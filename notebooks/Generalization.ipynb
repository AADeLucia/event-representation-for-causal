```python
# Greedy algorithm to match args with most frequent synsets across all args

def score_synset(synset_name, cur_cnt):
    # Only consider counts below `cur_cnt`, to align with the greedy nature of the approach
    return sum(cnt for _, cnt in inv_synset_index[synset_name] if cnt <= cur_cnt)

# First pass: build inverted index
inv_synset_index = collections.defaultdict(set)
for arg, cnt in common_args.most_common():
    if cnt >= 5:
        synsets = wn.synsets(arg, pos=wn.NOUN)
        for synset in synsets:
            parent_synsets = [y.name() for y in synset.hypernyms()]
            inv_synset_index[synset.name()].add((arg, cnt))
            for parent_synset in parent_synsets:
                inv_synset_index[parent_synset].add((arg, cnt))
        
# Second pass: assign args to synsets in a greedy manner
arg_to_synset = {}
for arg, cur_cnt in common_args.most_common(): 
    if arg not in arg_to_synset:
        synsets = wn.synsets(arg, pos=wn.NOUN)
        cand_synsets = []
        for synset in synsets:
            cand_synsets.append((score_synset(synset.name(), cur_cnt), 1, synset.name()))
            for parent_synset in synset.hypernyms():
                cand_synsets.append((score_synset(parent_synset.name(), cur_cnt), 0, parent_synset.name()))
        cand_synsets = list(set(cand_synsets))
        _, _, best_synset_name = max(cand_synsets)
        # Assign less-or-equal frequent args' synset to `best_synset_name`
        for lf_arg, cnt in inv_synset_index[best_synset_name]:
            if cnt <= cur_cnt:
                arg_to_synset[lf_arg] = best_synset_name
    # print(f'{arg} (cnt = {cur_cnt}) -> {arg_to_synset[arg]}')
```
